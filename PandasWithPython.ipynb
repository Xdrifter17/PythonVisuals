```python
#Merging the dataframes:


```

```python
df1=pd.read_csv(r"C:\Users\SAMAD\Downloads\Datasets\LOTR.csv")
df1
```

```python
df2=pd.read_csv(r"C:\Users\SAMAD\Downloads\Datasets\LOTR 2.csv")
df2
```

```python
#inner merge(which is default merge if not mentioned.)
df1.merge(df2)
```

```python
#outer merge
#NaN=Not a number(when there is no data)
df1.merge(df2, how = 'outer')
```

```python
#Left merge
df1.merge(df2, how = 'left')
```

```python
#Right merge
df1.merge(df2, how = 'right')
```

```python
df1.join(df2, on='FellowshipID', how = 'outer', lsuffix='left', rsuffix='right')
```

```python
#setting index
df4=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left', rsuffix='_right',how='outer')
df4
```

```python
pd.concat([df1,df2],join='outer',axis=1)
```

```python
#Data Cleaning with Pandas:


```

```python
df=pd.read_excel(r"C:\Users\SAMAD\Downloads\Datasets\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
# df["Last_Name"]=df["Last_Name"].str.strip("123._/")-fastest way to do the cleaning.
df["Last_Name"]=df["Last_Name"].str.lstrip("...")
df["Last_Name"]=df["Last_Name"].str.lstrip("/")
df["Last_Name"]=df["Last_Name"].str.rstrip("_")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
df[["Street_Address", "State","Zip"]]=df["Address"].str.split(',',2,expand=True)
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"]=='Y':
        df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"]=='':
        df.drop(x,inplace=True)
df
```

```python
df=df.reset_index(drop=True)
df
```
